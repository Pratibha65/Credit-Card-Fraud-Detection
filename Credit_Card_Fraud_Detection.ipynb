```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/creditcard.csv")
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.describe()
```

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit_sample  = legit.sample(n=104)
```

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
X=new_data.drop(['Class'],axis=1)
Y=new_data['Class']
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_test)
print(X_train)
```

```python
regressor = LogisticRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
training_prediction = regressor.predict(X_train)
training_accuracy = accuracy_score(training_prediction,Y_train)
print("Accuracy on training data : ",training_accuracy)
```

```python
testing_prediction = regressor.predict(X_test)
testing_accuracy = accuracy_score(testing_prediction,Y_test)
print("Accuracy on testing data : ",testing_accuracy)
```
